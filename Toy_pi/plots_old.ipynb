```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
%matplotlib notebook
```

```python
f = h5py.File('data_serial.hdf5', 'r')
```

```python
f['Domain/x'].shape
```

```python
print(f['Primitive/vx'].shape)
```

```python
print(f['Primitive/vy'][:])
print(np.all(np.isnan(f['Primitive/vy'][:])))
```

```python
plt.figure()
plt.plot(f['Domain/x'], f['Primitive/vy'][:])
plt.show()
```

```python
from scipy.special import erf
eta = f.attrs['gamma'][0]
end_t = f.attrs['t'][0]
tau_pi = f.attrs['sigma'][0]
print(eta, end_t, tau_pi)
plt.figure()
plt.plot(f['Domain/x'], 0.1*erf(f['Domain/x'][:]/(np.sqrt(4*eta*end_t))),label='theory')
plt.plot(f['Domain/x'], f['Primitive/vy'][:],label='numer')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(f['Domain/x'], f['Auxiliary/sigmaxy'][:])
plt.show()
```

```python
plt.figure()
plt.plot(f['Domain/x'], f['Auxiliary/Theta'][:])
plt.show()
```

```python
dx = f['Domain/x'][1]-f['Domain/x'][0]
dxT = (f['Primitive/T'][2:, 2] - f['Primitive/T'][:-2,2])/(2*dx)
plt.figure()
plt.plot(f['Domain/x'][1:-1], dxT)
plt.show()
```
