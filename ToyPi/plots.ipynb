```python
f = h5py.File('data_serial.hdf5', 'r')
f_id = h5py.File('Ideal/data_serial.hdf5', 'r')
```

```python
plt.figure()
plt.plot(f['Domain/x'], f['Primitive/vx'][:],label='Non-Ideal')
plt.plot(f['Domain/x'], f_id['Primitive/vx'][:],label='Ideal')
plt.title('Bulk viscous damping of a uniform fluid flow')
plt.xlabel(r'$x$')
plt.ylabel(r'$v_x$')
plt.xlim(0,1)
#plt.vlines(0.7+0.2*0.7+np.sqrt(0.1),0.0,0.7)
#plt.vlines(0.3+0.2*0.7,0.0,0.7)
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('toyPi_uniform_flow_vx.pdf')
```

```python
plt.figure()
plt.plot(f['Domain/x'], f['Primitive/Pi'][:],label='Non-Ideal')
plt.plot(f['Domain/x'], f_id['Primitive/Pi'][:],label='Ideal')
plt.xlabel(r'$x$')
plt.ylabel(r'$\Pi$')
plt.xlim(0,1)
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('toyPi_uniform_flow_Pi.pdf')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(6,6))
ax1.plot(f['Domain/x'], f['Primitive/vx'][:],label='Non-Ideal')
ax1.plot(f['Domain/x'], f_id['Primitive/vx'][:],label='Ideal')
#ax1.set_title('Bulk viscous damping of a uniform fluid flow')
#ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$v_x$')
ax1.set_xticklabels([])
ax1.set_xlim(0,1)
ax1.set_xticks([])
#plt.legend()
#plt.vlines(0.7+0.2*0.7+np.sqrt(0.1),0.0,0.7)
#plt.vlines(0.3+0.2*0.7,0.0,0.7)

ax2.plot(f['Domain/x'], f['Primitive/Pi'][:],label='Non-Ideal')
ax2.plot(f['Domain/x'], f_id['Primitive/Pi'][:],label='Ideal')
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$\Pi$')
ax2.set_xlim(0,1)

plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('toyPi_uniform_flow_both_vert.pdf')
```
