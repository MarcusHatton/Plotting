```python
def TwoDAnimation(i, VarClass, Var, num_frames):
    fs = fss[i]
    fig, ax = plt.subplots()
    images = []
    for frame_num in range(num_frames):
        images.append([plt.imshow(fs[frame_num][VarClass+'/'+Var][:,:])])

    anim = ArtistAnimation(fig, images, blit=True, repeat_delay=0, interval=500)
    return anim
```

```python
def TwoDAnimationDiff(i, j, VarClass, Var, num_frames):
    fs1 = fss[i]
    fs2 = fss[j]
    fig, ax = plt.subplots()
    images = []
    for frame_num in range(num_frames):
        images.append([plt.imshow((fs1[frame_num][VarClass+'/'+Var][:,:] - fs2[frame_num][VarClass+'/'+Var][:,:])/fs1[frame_num][VarClass+'/'+Var][:,:])])

    anim = ArtistAnimation(fig, images, blit=True, repeat_delay=0, interval=500)
    plt.colorbar()
    plt.clim(0,1)
    return anim
```
