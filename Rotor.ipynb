```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
```

```python
%matplotlib inline
```

```python
fss = []
for n in range(6):
  fss.append(h5py.File('ISCE/Rotor/highres/data_serial_'+str(n)+'.hdf5','r'))
```

```python
fig, axes = plt.subplots(2,3,figsize=(12,10))
for fs, axis in zip(fss,axes.flatten()):
    axis.imshow(fs['Primitive/n'])
plt.savefig('rotor_lowres.pdf')
```
