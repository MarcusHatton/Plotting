```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
```

```python
%matplotlib inline
```

```python
fss = []
for n in range(18):
  fss.append(h5py.File('ISCE/Rotor/data_serial_'+str(n)+'.hdf5','r'))
```

```python
fig, axes = plt.subplots(6,3,figsize=(10,20))
fig.tight_layout()
for fs, axis in zip(fss,axes.flatten()):
#     fig.tight_layout()
    axis.imshow(fs['Primitive/n'])
fig.subplots_adjust(hspace=0.2)
fig.subplots_adjust(wspace=0.2)  
plt.savefig('rotor_highres.pdf')
plt.show()
```
