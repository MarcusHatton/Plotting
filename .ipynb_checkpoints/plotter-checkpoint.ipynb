```python
fs1 = []
fs2 = []
fs3 = []
for n in range(6):
  fs1.append(h5py.File('IS/Shocktube1DPara/Ideal/data_serial_TIx_'+str(n)+'.hdf5','r'))
  fs2.append(h5py.File('IS/Shocktube1DPara/Bulk/data_serial_TIx_'+str(n)+'.hdf5','r'))
  fs3.append(h5py.File('ISCE/Shocktube1DPara/Bulk/data_serial_TIx_'+str(n)+'.hdf5','r'))
fss = [fs1, fs2, fs3]
```

```python
#print(fs_id[0]['Domain'].attrs['dt'])
for fs in fss:
    print(fs[0]['Optional'].attrs['zeta'])
```

```python
for fs in fss:
    plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Auxiliary/Theta'][:])
```

```python
print("INITIAL DATA")
for fs in fss:
    plt.plot(fs[0]['Domain/x'][:], fs[0]['Primitive/v1'][:], label='IS')
plt.legend()
plt.show()
for fs in fss:
    plt.plot(fs[0]['Domain/x'][:], fs[0]['Primitive/v2'][:], label='IS')
plt.legend()
plt.show()
```

```python
plt.figure
for fs in fss:
    plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Primitive/v1'][:], label='ISCE Ideal')
plt.title('v1')
plt.legend()
plt.savefig('v1.pdf')
plt.show()
```

```python
for fs in fss:
    plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Primitive/v2'][:], label='IS')
plt.legend()
plt.show()
```

```python
for var in 'D', 'S1', 'S2', 'Tau':
    for fs in fss:
        plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Conserved/'+var][:],label='ISCE Ideal')
    plt.title(var)
    plt.legend()
    plt.savefig(var+'.pdf')
    plt.show()
```

```python
for var in 'p', 'rho', 'n':
    for fs in fss:
        plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Primitive/'+var][:],label='ISCE Ideal')
    plt.title(var)
    plt.legend()
    plt.savefig(var+'.pdf')
    plt.show()
```

```python
for var in 'q1', 'q2', 'Pi', 'pi11', 'pi12', 'pi22':
    for fs in fss:
        plt.plot(fs[-1]['Domain/x'][:], fs[-1]['Auxiliary/'+var+'NS'][:],label=var+'NS')
    plt.legend()
    plt.show()
```
